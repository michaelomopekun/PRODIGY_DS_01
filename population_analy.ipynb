### Data Cleaning and Inspection

```python
df.info()
```

```python
print(f"This dataset has {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# check for duplicated values
print(f"Duplicated Values: {df.duplicated().any()}")
```

```python
# modifying the year column
filter=["[YR1990]","[YR2000]","[YR2014]","[YR2015]","[YR2016]","[YR2017]","[YR2018]","[YR2019]","[YR2020]","[YR2021]","[YR2022]","[YR2023]"]
for col in filter:
    df.columns=df.columns.str.replace(col,"")
df.columns
```

```python
# check for null values
df.isnull().sum()
```

```python
#preview rows with null value
df[df["Country Name"].isnull()]
```

```python
#droping rows with null values
df.dropna(inplace=True)
df.isnull().sum()
```

```python
# check for unique values 
df.nunique()
```

###### droping the series name,series code and country code, since we wont be needing them in our analysis 

```python
# drop columns
df.drop(columns=["Series Name","Series Code","Country Code"],inplace=True)
df.head(10)
```

```python
# converting type to float
for col in df.columns[1:]:
    df[col]=pd.to_numeric(df[col],errors='coerce')
df.dtypes
```

```python
# slicing into two categories
cont_df=df.iloc[:217]
cat_df=df.iloc[217:]
cat_df.columns
```

### Visualizing

```python
# world population over the years
cat_df_world_pop=cat_df[cat_df["Country Name"]=="World"].set_index('Country Name').to_dict()
pop_overview=pd.DataFrame.from_dict(cat_df_world_pop,orient='index')
pop_overview.index.name='Year'
pop_overview.reset_index(inplace=True)
pop_overview
```

```python
plt.figure(figsize=(10,7))
sns.barplot(data=pop_overview,x='World',y='Year')
plt.xlabel('World_population')
plt.title('World Population over the years')
plt.show()
```

```python
# top 10 most populated country
top_10_populated=cont_df.sort_values(by="2023 ",ascending=False).head(10).set_index('Country Name').to_dict()
top_10_populated=pd.DataFrame.from_dict(top_10_populated,orient='index')
top_10_populated.index.name='Year'
top_10_populated.reset_index(inplace=True)
top_10_populated
```

```python

column_inTop10=top_10_populated.columns
fig,axes=plt.subplots((len(column_inTop10)-1),1,figsize=(9,60))
y_column='Year'
# visualizing each countries population over time
for i,col in enumerate(column_inTop10[1:]):
    sns.barplot(data=top_10_populated,x=col,y=y_column,ax=axes[i])
    axes[i].set_title(f'{col} Population over the years')
    
plt.tight_layout() 
plt.show() 
```

```python
# lowest_populated country
least_10_populated=cont_df.sort_values(by="2023 ",ascending=False).tail(10).set_index('Country Name').to_dict()
least_10_populated=pd.DataFrame.from_dict(least_10_populated,orient='index')
least_10_populated.index.name='Year'
least_10_populated.reset_index(inplace=True)
least_10_populated
```

```python
column_inleast10=least_10_populated.columns
fig,axes=plt.subplots((len(column_inleast10)-1),1,figsize=(9,60))
y_column='Year'
# visualizing each countries population over time
for i,col in enumerate(column_inleast10[1:]):
    sns.barplot(data=least_10_populated,x=col,y=y_column,ax=axes[i])
    axes[i].set_title(f'{col} Population over the years')
    
plt.tight_layout() 
plt.show() 
```
